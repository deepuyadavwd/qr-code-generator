```python
import qrcode
from IPython.display import Image, display

# Define the fields as per the form
fields = [
    "HR Coil No.",
    "Batch No.",
    "Grade",
    "Heat No.",
    "Only Thickness",
    "Finish",
    "Edge Condition",
    "Quality",
    "QA Inspector",
    "Work Center",
    "Sample Location",
    "Remarks"
]

# Collect data from user input
data = {}
for field in fields:
    data[field] = input(f"Enter value for {field}: ")

# Generate the data string for the QR code
qr_data = "\n".join([f"{field}: {value}" for field, value in data.items()])

# Create and save the QR code
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data(qr_data)
qr.make(fit=True)
img = qr.make_image(fill_color="black", back_color="white")
img.save("qr_label.png")

# Print the formatted label
print("\nQR Code Label\n")
print("{:<20} {:<}".format("Field Name", "Value"))
print("-" * 40)
for field in fields:
    print("{:<20} {:<}".format(field, data[field]))
print("\nQR code saved as qr_label.png")

# Display the QR code image in Google Colab
display(Image(filename="qr_label.png"))
```

```python
from flask import Flask, render_template, request
import qrcode
import base64
from io import BytesIO

app = Flask(__name__)

@app.route("/", methods=["GET", "POST"])
def index():
    qr_code = None
    if request.method == "POST":
        # Define the fields as per the form
        fields = [
            "HR Col No.",
            "Batch No.",
            "Grade",
            "Heat No.",
            "Only Thickness",
            "Finish",
            "Edge Condition",
            "Quality",
            "QA Inspector",
            "Work Center",
            "Sample Location",
            "Remarks"
        ]

        # Collect data from the form
        data = {field: request.form[field] for field in fields}

        # Generate the data string for the QR code
        qr_data = "\n".join([f"{field}: {value}" for field, value in data.items()])

        # Create the QR code
        qr = qrcode.QRCode(
            version=1,
            error_correction=qrcode.constants.ERROR_CORRECT_L,
            box_size=10,
            border=4,
        )
        qr.add_data(qr_data)
        qr.make(fit=True)

        # Convert the QR code image to a base64 string
        img = qr.make_image(fill_color="black", back_color="white")
        buffered = BytesIO()
        img.save(buffered, format="PNG")
        qr_code = base64.b64encode(buffered.getvalue()).decode()

    return render_template("index.html", qr_code=qr_code)

if __name__ == "__main__":
    app.run(debug=True)
```
